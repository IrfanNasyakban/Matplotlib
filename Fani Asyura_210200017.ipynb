## A. Membuat Visualisasi Data dengan Line Plots

```python
import pandas as                   # mengimport panda
import numpy as np                 # mengimport numpy
%matplotlib inline                 # gunakan %matplotlib inline agar hasil visualisasi bisa tercetak di Jupyter Notebook
import matplotlib as mpl           # mengimport matplotlib
import matplotlib.pyplot as plt    # mengimport matplotlib.pyplot

# Membaca file Canada.xlsx, dan mengedit nya
df = pd.read_excel('Canada.xlsx', sheet_name='Canada by Citizenship', skiprows=range(20), skipfooter=2)
df.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
df.rename(columns={'OdName':'Negara', 'AreaName':'Benua', 'RegName':'Wilayah', 'DevName':'Kemajuan'}, inplace=True)
df.set_index('Negara', inplace=True)
df.index.name = None
df.columns = list(map(str, df.columns))

# Membuat list kolom tahun 1980 - 2013
tahun = list(map(str, range(1980, 2013)))
Indonesia = df.loc['Indonesia', tahun]
Indonesia.index = Indonesia.index.map(int)
mpl.style.use(['ggplot'])

# Membuat line plot
Indonesia.plot(kind='area')
plt.title('Jumlah imigran dari Indonesia yang ada di kanada')
plt.ylabel('Jumlah imigran')
plt.xlabel('Tahun')
plt.text(2000, 1100, 'lonjakan sangat tinggi')

# Menampilkan line plot
plt.show()
```

## B. Membuat Visualisasi Data dengan Area Plots

```python
# Membuat Area Plot
imigran    = {"China":df.loc['China', tahun],

               "India":df.loc['India', tahun]};

# Membuat instance DataFrame

dataFrame   = pd.DataFrame(imigran, index=tahun);

#Gambarkan plot area untuk data DataFrame

dataFrame.plot(kind='area', stacked=False)
plt.title('Perbandingan jumlah imigran China dan India di Canada')
plt.ylabel('Jumlah Imigran')
plt.xlabel('Tahun')
plt.show(block=True);
```

## C. Membuat Visualisasi Data dengan Histogram

```python
# Membuat Histogram
df['2000'].head(10).plot(kind='hist', figsize=(5,5))
plt.title('Jumlah Negara tahun 2000')
plt.ylabel('Jumlah Negara')
plt.xlabel('Jumlah Imigran')

# Menampilkan Histogram
plt.show(close=None, block=None)
```

## D. Membuat Visualisasi Data dengan Bar Plots

```python
# Membuat Bar plot
objek = list(map(str, range(1980, 2013)))

height = df.loc['China', tahun]
height.index = height.index.map(int)

fig = plt.figure(figsize = (9, 4))

plt.bar(objek, height)

plt.xlabel("Tahun")
plt.xticks(rotation=90)
plt.ylabel("Jumlah imigran")
plt.title("Jumlah Imigran dari China setiap tahunnya")
plt.show()
```

## E. Membuat Visualisasi Data dengan Grafik Pie

```python
# Membuat grafik pie

benua = ['Malaysia', 'China', 'Indonesia', 'India']
sizes = [40, 30, 10,70]

fig, ax = plt.subplots()
ax.pie(sizes, labels=benua, autopct='%1.1f%%')
ax.axis('equal')
ax.set_title('pie jumlah imigran menurut kelompok Benua khusus tahun 2000')
plt.show(close=None, block=None)
```

## F. Membuat Visualisasi Data dengan Box Plots

```python
# Membuat grafik box plots
India = df.loc['India', tahun]
India.index = India.index.map(int)
mpl.style.use(['ggplot'])

India.plot(kind='box')
plt.title('Jumlah imigran dari India')
plt.ylabel('Jumlah imigran')
plt.xlabel('Negara')
plt.show(close=None, block=None)
```
